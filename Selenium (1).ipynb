```python
driver.get("https://www.flipkart.com")

obj = driver.find_element(By.CLASS_NAME,'_3704LK')
obj.send_keys("sunglasses")

obj = driver.find_element(By.CLASS_NAME, 'L0Z3Pu')
obj.click()
```

```python
count = 0
all_brands = []
all_des = []
all_prices = []

for i in range(3):
    brands = driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in brands:
        all_brands.append(i.text)

    des = driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for i in des:
        all_des.append(i.text)

    prices = driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for i in prices:
        all_prices.append(i.text)

    count += len(brands)
    obj = driver.find_elements(By.CLASS_NAME, "_1LKTO3")
    if len(obj)>1:
        obj[1].click()
    else:
        obj[0].click()
    time.sleep(5)
    
    
data = {'Brand': all_brands[:100], 'Description': all_des[:100], 'Price': all_prices[:100]}
df = pd.DataFrame(data)
df
```

```python
all_brands = []
all_des = []
all_prices = []

for i in range(2):
    brands = driver.find_elements(By.XPATH,'//h3[@class="product-brand"]')
    for i in brands:
        all_brands.append(i.text)

    des = driver.find_elements(By.XPATH,'//h4[@class="product-product"]')
    for i in des:
        all_des.append(i.text)

    prices = driver.find_elements(By.XPATH,'//div[@class="product-price"]')
    for i in prices:
        all_prices.append('Rs' + i.text.split('Rs')[1])

    count += len(brands)
    obj = driver.find_elements(By.CLASS_NAME, "pagination-next")
    if len(obj)>1:
        obj[1].click()
    else:
        obj[0].click()
    time.sleep(5)
    
    
data = {'Brand': all_brands[:100], 'Description': all_des[:100], 'Price': all_prices[:100]}
df = pd.DataFrame(data)
df
```

```python
obj = driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[1]/div[2]/div/div[3]/span/div[1]/div/div/div[6]/ul[4]/li[10]/span')
obj.click()
```

```python


all_brands = []
all_des = []
all_prices = []


brands = driver.find_elements(By.XPATH,'//span[@class="a-size-medium a-color-base a-text-normal"]')
for i in brands:
    all_brands.append(i.text)

des = driver.find_elements(By.XPATH,'//div[@class="a-row a-size-small"]')
for i in des:
    x = i.find_elements(By.TAG_NAME, 'span')
    all_des.append(x[0].get_attribute('aria-label'))

prices = driver.find_elements(By.XPATH,'//span[@class="a-price-whole"]')
for i in prices:
    all_prices.append(i.text)

data = {'Name': all_brands[:10], 'Rating': all_des[:10], 'Price': all_prices[:10]}
df = pd.DataFrame(data)
df
  
```
